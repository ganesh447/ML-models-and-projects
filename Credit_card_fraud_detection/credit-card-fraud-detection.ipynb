## Dealing with unbalanced data
99.9% data is legit and very small amount of data is fraudulent  
**Under-Sampling Technique**

#Build a sample dataset containing simillar distribution of normal transaction(legit) and fraudelent transaction  
- Number of fraudelent transaction is 492
- Create a sample dataset from legit transaction using **Random Sampling** and join it with fraudelent dataset to make uniform distribution

```python
legit_sample =legit.sample(n=492)
legit_sample.shape
```

concatenating both data frame  
- legit_sample (492, 31)
- fraud(492, 31)

```python
new_df = pd.concat([legit_sample,fraud], axis= 0)
print(new_df.shape)
print(new_df.head())
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class')['Amount'].mean().round(3)
```

**Observation**
- mean value is similar to the population mean value

<h1> Feature Engineering </h1>

```python
#importing model and train_test_split

from sklearn.model_selection import train_test_split   #importing train_test_split for spliting data
from sklearn.linear_model import LogisticRegression    #importing LogisticRegression as the problem is classification problem
from sklearn.metrics import accuracy_score             #importing performance matrics accury_score to check model accuracy
```

```python
X = new_df.drop(columns = 'Class', axis = 1)
Y= new_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)  #spliting randomly
```

```python
print('Shape of X_train', X_train.shape)
print('Shape of Y_train', Y_train.shape)
print('Shape of X_test', X_test.shape)
print('Shape of Y_test', Y_test.shape)
```

<h3> We use logistic regression as we have a classification problem </h3>

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#Model Evaluation
X_train_predict = model.predict(X_train)
Training_data_accurary = accuracy_score(X_train_predict, Y_train)
Training_data_accurary

#printing accuracy on training data
print("Training Data Accuracy is", Training_data_accurary)
```

```python
X_test_pred = model.predict(X_test)
Testing_data_accuracy = accuracy_score(X_test_pred, Y_test)
print("Testing Data Accuracy is ", Testing_data_accuracy)
```

```python
results = pd.DataFrame({
    "Actual": Y_test,
    "Predicted": X_test_pred
})
```
